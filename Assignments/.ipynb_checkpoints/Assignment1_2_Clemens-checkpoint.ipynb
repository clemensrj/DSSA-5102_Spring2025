Make all column names lowercase, with no spaces or special characters:

```python
# Make column names lowercase, remove spaces, and remove all special characters from the column names for easier compatibility in SQL
df_1985_2022.columns = df_1985_2022.columns.str.lower().str.replace(' ', '').str.replace(r'[^\w\s]', '', regex=True)
```

```python
# Check for missing values and print a summary
print("\nMissing values in each column:\n", df_1985_2022.isnull().sum())
```

```python
# Handle missing values in 'Show_name' (replace with 'Unknown')
df_1985_2022['show_name'] = df_1985_2022['show_name'].fillna('Unknown') #This modification will help readers understand why there is no show name, the data is unknown
```

```python
# Print the rows where 'Potential_Gross' is 0
zero_potential_gross = df_1985_2022[df_1985_2022['potential_gross'] == 0]

# Display the rows with 0 'Potential_Gross'
print("\nRows with 0 'Potential_Gross':\n", zero_potential_gross)
```

```python
# Print the modified column names to verify the changes
print("\nModified column names:\n", df_1985_2022.columns)
```

```python
# Drop completely duplicated rows (if they exist) 
df_1985_2022.drop_duplicates(inplace=True)
#If we remove completely duplicated rows we can cut down on entries to show better generalized trends for specific shows or years. However, we need to keep duplicate entries per column like show name because the set represents performances of shows per week so there will be many valid duplicates. 
```
