```python
# Assignment 1
# Name: Robert Clemens
# Date: Febuary 6, 2025
# Course: DSSA 5201

# Import libraries
import pandas as pd #Used for data manipulation and analyzing
import numpy as np #Used to handle numerical data and functions
import matplotlib.pyplot as plt #Used to create scatter plot, box plot, and other plots
import seaborn as sns #used to create graphs and other figures

# Load the dataset
# If the data is stored locally
penguins = pd.read_csv("penguins_clean.csv")

# Display the first few rows to understand the structure of the data loaded in.
print("Beginning of Data Set:")
print(penguins.head())

# Display info about the types of rows and columns used. 
print("Types of Columns and Rows in Dataset:")
print(penguins.info())

# Extract column names from the data 
print("Column Names:")
print(penguins.columns.tolist())

# Identify the unique species that are under the Species column
print("Penguin Species:")
print(penguins["species"].unique())

# Identify the unique islands that are listed under the island column
print("Islands:")
print(penguins["island"].unique())

# Count the number of penguins per species
print("Number of Penguins per Species:")
print(penguins["species"].value_counts())

# Count the number of penguins per island
print("Number of Penguins per Island:")
print(penguins["island"].value_counts())

# Check for any missing data within the data set
print("Missing Data:")
print(penguins.isnull().sum())

# Summary of information about the numerical columns
print("Summary Statistics:")
print(penguins.describe())

# This creates a box plot which will help visualize the outliers within the data
plt.figure(figsize=(10, 6))
sns.boxplot(data=penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']])
plt.title("Boxplots of Multiple Features")
plt.show()

# This creates a scatter plot of flipper length vs. bodymass with species separated by color
plt.figure(figsize=(8, 6))
sns.scatterplot(x=penguins['flipper_length_mm'], y=penguins['body_mass_g'], hue=penguins['species'])
plt.title("Flipper Length vs Body Mass by Species")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.show()

# Summary of Findings:
# - The dataset contains three species of penguins: Gentoo (124), Adelie (152), and Chinstrap (68).
# - Penguins are on three islands: Biscoe (168), Dream (124), and Torgersen (52).
# - There are missing values in a few columns: bill_length_mm (2), bill_depth_mm (2), flipper_length_mm (2), body_mass_g (2), and sex (11).
# - The avg bill length is approximately 43.92 mm, with a range from 32.1 mm to 59.6 mm.
# - The avg body mass is 4201.75 grams, ranging from 2700 g to 6300 g.
# - Boxplots indicate potential outliers in body mass and bill measurements.
# - A scatter plot of flipper length vs. body mass reveals a distinct separation between species.
# - The dataset spans the years 2007 to 2009, with most data points clustered around 2008.
# - Overall, the dataset provides a well-structured view of the Palmer penguin species, with some missing data requiring attention.

# End of script
```
